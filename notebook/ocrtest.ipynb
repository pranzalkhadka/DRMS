```python
import pytesseract
from PIL import Image
```

## Draw bounding box and print text

```python
from PIL import ImageDraw

image_path2 = r"D:\custom_dataset\ocr all\17.jpg"
```

```python

def ocr_exec(model,filepath):
    image = Image.open(filepath)
    text = pytesseract.image_to_string(image, lang='nep')
    boxes = pytesseract.image_to_boxes(image, lang=model)

    # Draw bounding boxes on the image
    draw = ImageDraw.Draw(image)
    width, height = image.size

    for box in boxes.splitlines():
        b = box.split(' ')
        draw.rectangle([int(b[1]), height - int(b[2]), int(b[3]), height - int(b[4])], outline='red')

  
    print(text)
    image.show()  # To display the image

   
```

```python
ocr_exec('nep',image_path2)
```

```python
ocr_exec('nep-1200',image_path2)
```

## Generate PDF

```python
# Load the image using PIL
image = Image.open(image_path2)

# Get a searchable PDF
pdf = pytesseract.image_to_pdf_or_hocr(image, extension='pdf',lang='nep')
with open('test.pdf', 'w+b') as f:
    f.write(pdf) # pdf type is bytes by default
```
