```python
template_path = r"D:\Document-Management-of-Nepali-Papers\dataset\data\unprepared\ID\master"
output_folder = r"D:\Document-Management-of-Nepali-Papers\dataset\data\unprepared\ID\master\output"
```

```python
variables_files = {
    'district': 'districts.txt',
    'name': 'names.txt',
    'gender': 'gender.txt',
    'VDC': 'vdc.txt',
    'ward': 'ward.txt',
    'year': 'year.txt',
    'month': 'month.txt',
    'number': 'number.txt',
    'parent1': 'parent1.txt',
    'parent2': 'parent2.txt',
    'district2':'district2.txt',
    'ward2':'ward2.txt',
    'ward3':'ward3.txt',
    'VDC2':'vdc2.txt',
}
```

```python
# Read the template
with open(f"{template_path}/template.txt", 'r', encoding='utf-8') as template_file:
    template = template_file.read()
```

```python
template
```

```python
variables = {}
for key, file in variables_files.items():
    file_path= f"{template_path}\{file}"
    with open(file_path, 'r', encoding='utf-8') as var_file:
        variables[key] = var_file.read().strip().split(',')
```

```python
variables
```

```python
# Generate 100 text files
for i in range(500):
    # Replace placeholders in the template
    content = template
    for key, values in variables.items():
        placeholder = f'({key})'
        value = values[i % len(values)]
        content = content.replace(placeholder, value)

    # Create the output file
    output_path = os.path.join(output_folder, f'{i+1}.txt')
    with open(output_path, 'w', encoding='utf-8') as output_file:
        output_file.write(content)

print("Text files generated successfully.")
```
