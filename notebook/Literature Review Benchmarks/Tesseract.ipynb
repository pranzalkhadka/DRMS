```python
import pytesseract
from PIL import Image
```

```python
import csv
import time
from docx import Document
```

```python
doc = Document()
csv_file_path = "/content/OCR TEST/dataup.csv"
image_directory = "/content/OCR TEST/dataset1"
```

```python
#Testing file data
with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    # Create a CSV reader object
    csv_reader = csv.reader(csvfile)

    # Skip the header row if your CSV has one
    next(csv_reader, None)

    # Loop through each row in the CSV
    for row in csv_reader:
        print(row)
```

##MAIN LOOP

```python
import os
import jiwer
import numpy as np
cers = []
wers = []
latencies=[]
doc = Document()
for i in range(1, 60):
    image_path = os.path.join(image_directory, f'{i}.png')

    # Measure start time
    start_time = time.time()

    # Perform OCR on the image
    image = Image.open(image_path)
    ocr_text = pytesseract.image_to_string(image, lang='nep')
    print(ocr_text)
    # Measure end time
    end_time = time.time()

    # Calculate the number of words in the OCR result
    num_words = len(ocr_text.split())

    # Calculate the total latency and average latency per word
    total_latency = end_time - start_time
    avg_latency_per_word = total_latency / num_words if num_words else 0
    latencies.append(avg_latency_per_word)
    with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
      # Create a CSV reader object
      csv_reader = csv.reader(csvfile)
      for row_number, row in enumerate(csv_reader, start=1):
            # Assuming each row corresponds to each image in order
            if row_number == i:
                reference_text = row[0]
                break

    cer = jiwer.cer(reference_text, ocr_text)
    wer = jiwer.wer(reference_text, ocr_text)
    cers.append(cer)
    wers.append(wer)
    # doc.add_paragraph(f"Image Path: {image_path}\nOCR Text: {ocr_text}\nCER: {cer}\nWER: {wer}\nLatency per word: {avg_latency_per_word} seconds")
    print(f"Image: {i}.png, CER:{cer} | WER:{wer} \nAverage Latency per Word: {avg_latency_per_word:.4f} seconds\n")

doc.save('ocr_results.docx')
average_cer = np.mean(cers)
average_wer = np.mean(wers)
average_latency = np.mean(latencies)
print(f"Average Latency: {average_latency:.4f} seconds")
print(f"Average CER: {average_cer:.2f}, Average WER: {average_wer:.2f}")
```
